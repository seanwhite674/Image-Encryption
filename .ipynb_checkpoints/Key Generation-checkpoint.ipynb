# Generating a key using the logistic map

```python
def keygen(x,r,size):
    key=[]
    for i in range(size):
        x=r*x*(1-x)   # logistic map
        key.append(int((x*pow(10,16))%256)) 
    return key      
```

```python
print(keygen(0.01,3.9394,10))
print(keygen(0.001,3.9394,10))
print(keygen(0.001,3.95,10))
```

# Shuffling elements of an array part one

```python
import numpy as np
a=[10,20,30,40,50]
index=[3,2,0,1,4]
print("Original array:",a)
b=[]
#shuffling
for i in range(5):
    b.append(a[index[i]])
print("Shuffled array:",b)
c=np.zeros(5,dtype=int)
#Reshuffling
for i in range(5):
    c[index[i]]=b[i]
print("Reshuffled array:",c)
```

# Shuffling a Matrix

```python
index=[1,0,2]
m=[[3,2,1],[5,4,6]]
m1=np.zeros([len(m),len(m[0])],dtype=int)
print("Original Matrix:",m)

#Matrix Shuffling
for i in range(len(m)):#height
    k=0
    for j in range(len(m[0])):#width
        m1[i][j]=m[i][index[k]]
        #print(i,"",index[k],"",m[i][index[k]])
        k=k+1
print("Shuffled Matrix:",m1)

#Matrix Re-shuffling
for i in range(len(m)):#height
    k=0
    for j in range(len(m[0])):#width
        m[i][index[k]]=m1[i][j]
        k=k+1
print("Unshuffled Matrix:",m)
        
```
